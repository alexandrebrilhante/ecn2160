## Série chronologique

```r
# On génère 12 valeurs aléatoires.

input <- runif(12)
```

```r
input
```

```r
# On convertit nos points en série chronologique.

input <- ts(input, start = c(2018, 1), frequency = 12)
```

```r
input
```

```r
plot.ts(input)
```
