## Série chronologique

## Régression

```r
# Création de données.
set.seed(123)

x <- 1:20 + rnorm(20, sd = 3)
z <- 1:20/4 + rnorm(20, sd = 2)
y <- -2*x + x*z/5 + 3 + rnorm(20, sd = 4)

# On créé un datarame avec nos données.
dat <- data.frame(x = x, y = y, z = z)

head(dat)
```

#### Corrélation

```r
cor(dat$x, dat$y)
```

#### Régression linéaire

```r
fit <- lm(dat$y ~ dat$x)

fit
```

Cela veut donc dire que y = -1.1786-0.8587*x.

```r
summary(fit)
```
